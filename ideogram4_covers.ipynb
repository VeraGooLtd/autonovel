```python
# Step 3: Load Ideogram 4
import torch, os, time
from diffusers import Ideogram4Pipeline

print('Downloading Ideogram 4 (nf4) model...')
print('This takes 10-20 minutes on first run...')

pipe = Ideogram4Pipeline.from_pretrained(
    'ideogram-ai/ideogram-4-nf4',
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map='auto',
    low_cpu_mem_usage=True,
)

print('Ideogram 4 loaded!')
print('Device:', pipe.device)
```

```python
# Step 4: Generate covers
OUTPUT = '/content/covers'
os.makedirs(OUTPUT, exist_ok=True)

covers = [
  ('the-genesis-engine', 'The Genesis Engine', 'Middle grade sci-fi book cover 1600x2400. Title The Genesis Engine bold futuristic top. Author Daniel Ryan bottom. Fungal spore pod on Mars Earth orange sky. Clean bold cartoonish.'),
  ('the-martian-spore', 'The Martian Spore', 'Middle grade sci-fi book cover 1600x2400. Title The Martian Spore bold futuristic top. Author Daniel Ryan bottom. Cracked Mars soil green mycelium. Clean bold cartoonish.'),
  ('the-crimson-bloom', 'The Crimson Bloom', 'Middle grade sci-fi book cover 1600x2400. Title The Crimson Bloom bold futuristic top. Author Daniel Ryan bottom. Crimson bloom Mars from space spiral. Clean bold cartoonish.'),
  ('sporefall', 'Sporefall', 'YA sci-fi horror book cover 1600x2400. Title Sporefall distressed grunge top. Author Daniel Ryan bottom. Fungal fruiting body dark rainforest. Dark atmospheric stylized.'),
  ('mycelial-tide', 'Mycelial Tide', 'YA sci-fi horror book cover 1600x2400. Title Mycelial Tide distressed grunge top. Author Daniel Ryan bottom. Mycelium wall city street. Dark atmospheric stylized.'),
  ('mycelial-earth', 'Mycelial Earth', 'YA sci-fi horror book cover 1600x2400. Title Mycelial Earth distressed grunge top. Author Daniel Ryan bottom. Earth space fungal web. Epic dark stylized.'),
  ('the-salami-saboteur', 'The Salami Saboteur', 'Middle grade mystery humor book cover 1600x2400. Title The Salami Saboteur playful handwritten top. Author Daniel Ryan bottom. Hand-drawn doodle Diary of a Wimpy Kid style. Ginger cat domino mask salami. Girl glasses bulldog shocked. White background.'),
  ('the-glass-vaults-gambit', 'The Glass Vaults Gambit', 'Middle grade mystery humor book cover 1600x2400. Title The Glass Vaults Gambit playful handwritten top. Author Daniel Ryan bottom. Hand-drawn doodle Diary of a Wimpy Kid style. Mall night laser grid. Ginger cat turtleneck air duct. White background.'),
  ('station-zero', 'Station Zero', 'YA sci-fi book cover 1600x2400. Title Station Zero bold futuristic top. Author Daniel Ryan bottom. Derelict space station deep space. Shuttle approaching. Clean sci-fi.'),
  ('the-echo-chamber', 'The Echo Chamber', 'YA sci-fi thriller book cover 1600x2400. Title The Echo Chamber bold tech-thriller top. Author Daniel Ryan bottom. Face phone AI avatar reflection. Glitch cyberpunk.'),
  ('the-silent-strata', 'The Silent Strata', 'YA sci-fi book cover 1600x2400. Title The Silent Strata bold futuristic top. Author Daniel Ryan bottom. Earth cross-section geological layers. Alien structure glowing deep. Clean sci-fi.'),
]

for i, (slug, title, prompt) in enumerate(covers, 1):
    print(f'[{i}/{len(covers)}] {title}...')
    try:
        img = pipe(prompt=prompt, width=1600, height=2400, num_inference_steps=30, guidance_scale=7.5).images[0]
        img.save(f'{OUTPUT}/{slug}-ebook-cover.png')
        print('  OK')
    except Exception as e:
        print(f'  FAIL: {e}')
print('Done!')
```
